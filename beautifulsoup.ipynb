```python
from bs4 import BeautifulSoup

html_code="""
<!DOCTYPE html>
<html>
<body>

<p>This is a paragraph.</p>

</body>
</html>
"""

html_content=BeautifulSoup(html_code, "html")

# print(html_content)

# print(type(html_content))
# print(html_content.text)

# print(html_content.find("p"))
print(html_content.find("p").text)
```

```python
html_code="""
<!DOCTYPE html>
<html>
<body>

<p>This is a paragraph.</p>
<p>This is another paragraph.</p>

</body>
</html>
"""

# BeautifulSoup(html_code, "html")
html_content=BeautifulSoup(html_code, "html")
# print(html_content.find_all("p"))

all_para=html_content.find_all("p")

for p in all_para:
  print(p)
```
