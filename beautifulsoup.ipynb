```python
response=requests.get(url)
html_content=BeautifulSoup(response.content, "html")

data=[]

quote_text=html_content.select('span.text')
# quote_text

for txt in quote_text:
  # print(txt.text)
  data.append(txt.text)

print(data)
```

```python
import pandas as pd

response=requests.get(url)

html_content=BeautifulSoup(response.content, "html.parser")

author_data=[]
text_data=[]

quote_author=html_content.select("small.author")
quote_text=html_content.select("span.text")

for i in quote_author:
  author_data.append(i.text)

for i in quote_text:
  text_data.append(i.text)

df=pd.DataFrame({"Author":author_data, "Qoute":text_data})
df
```
